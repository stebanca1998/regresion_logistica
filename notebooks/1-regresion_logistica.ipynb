## Regresion Logistica Binomial

### Carga de datos

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_data_processing = pd.read_csv('../data/processed/Telco-Customer-Churn_processed.csv')
df_data_visualize = pd.read_csv('../data/processed/Telco-Customer-Churn_visualization.csv')
```

### Analisis de correlacion

```python
fig = plt.figure(figsize=(15,9))
df_data_processing.corr()['Churn'].sort_values(ascending=True).plot(kind='bar')
plt.show()
```

### Escalar los datos

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df_data_processing_scaled = scaler.fit_transform(df_data_processing) 
df_data_processing_scaled = pd.DataFrame(df_data_processing_scaled, columns=df_data_processing.columns)
```

### Analisis exploratorio

```python
import seaborn as sns
```

```python
sns.countplot(data=df_data_visualize, x='gender', hue='Churn')
plt.show()
```

```python
def plot_categorial(column):
    fig = plt.figure(figsize=(10,10))
    sns.countplot(data=df_data_visualize, x=column,hue='Churn')
    plt.show()
```

```python
column_cat = df_data_visualize.select_dtypes(include='object').columns
```

```python
for _ in column_cat:
    plot_categorial(_)
```

```python
fig = plt.figure(figsize=(10,10))
sns.pairplot(data= df_data_visualize, hue='Churn')
plt.show()
```
