## Regresion Logistica Binomial

### Carga de datos

```python
import pandas as pd
import numpy as np
```

```python
df_data = pd.read_csv('../data/raw/Telco-Customer-Churn.csv')
```

```python
df_data.head()
```

```python
df_data.info()
```

### Cambiar total_charges a numeric

```python
df_data.TotalCharges = pd.to_numeric(df_data.TotalCharges, errors='coerce')
```

#### Manejo de datos nulos

```python
df_data.isnull().sum()
```

##### Eliminar datos nulos

```python
## No que no representan una cantidad considerable
## 11 de 7000+
df_data.dropna(inplace=True)
```

##### Eliminar columna customerID

```python
## No es una variable representativas más que un identificador de compañia
df_data.drop('customerID',axis=1,inplace=True)
```

##### Convertir Variable Objetivo a Numerico

```python
## Ahora mismo la variable objetivo es tipo texto
df_data['Churn'] = df_data['Churn'].replace('Yes', 1)
df_data['Churn'] = df_data['Churn'].replace('No', 0)
```

#### Manejo de variables categóricas

```python
df_data = pd.get_dummies(df_data, dtype=int)
df_data.head()
```

```python
df_data.to_csv('../data/processed/Telco-Customer-Churn_processed.csv', index=False)
```
